```
!pip install resnet_pytorch
import json
import torch
import torchvision.transforms as transforms
from PIL import Image
from resnet_pytorch import ResNet 
model = ResNet.from_pretrained('resnet18')
img_dir = "/content/gdrive/My Drive/dataset/nonmeme" 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
data=[]
entire_text=[]
sets=[]
img_features=[]
print(len(files))
for f in range(9000,9500):
    print(f)
    data.append(files[f])
    with open(files[f], 'rb') as image_file:
        content = image_file.read()
    image1 = vision.types.Image(content=content)
    text_response = client.text_detection(image=image1)
    texts = [text.description for text in text_response.text_annotations]
    text=" ".join(texts[1:])
    entire_text.append(text)
    print(text)
    input_image = Image.open(files[f])

    preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    input_tensor = preprocess(input_image)
    input_batch = input_tensor.unsqueeze(0)
    inputs = input_batch
    features = model.extract_features(inputs)
    features=features.tolist()
    f=[]
    for i in range(512):
      f.append(features[0][i][0][0])
    img_features.append(f)
    



img_dictionary = dict(zip(files, img_features))
print(img_dictionary)





```

```
import torch
from pytorch_pretrained_bert import BertTokenizer, BertModel, BertForMaskedLM


import logging


import matplotlib.pyplot as plt
% matplotlib inline


tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```
